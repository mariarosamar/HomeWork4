```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
```

```python
# Reading whale returns
whale_returns_csv = Path("whale_returns.csv")
```

```python
# Reading the file
whale_returns_dataframe = pd.read_csv(whale_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
whale_returns_dataframe.head(5)
```

```python
# Count nulls
whale_returns_dataframe.isnull()
whale_returns_dataframe.isnull().sum()
```

```python
# Determining percentage of nulls
whale_returns_dataframe.isnull().mean() * 100
```

```python
# Drop nulls
whale_returns_dataframe.dropna(inplace=True)
#checking number of nulls
whale_returns_dataframe.isnull().sum()
```

```python
# Algorithmic Daily Returns
```

```python
# Reading algorithmic returns
algo_returns_csv = Path("algo_returns.csv")
# Reading the file
algo_returns_dataframe = pd.read_csv(algo_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo_returns_dataframe.head(5)
```

```python
# Count nulls
algo_returns_dataframe.isnull()
algo_returns_dataframe.isnull().sum()
```

```python
# Drop nulls
algo_returns_dataframe.dropna(inplace=True)
algo_returns_dataframe.isnull().sum()
```

```python
# S&P 500 Returns

# Reading S&P 500 Closing Prices
sp500_history_csv = Path("sp500_history.csv")
sp500_history_dataframe = pd.read_csv(sp500_history_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp500_history_dataframe.head(5)
```

```python
# Check Data Types
sp500_history_dataframe.dtypes
```

```python
# Fix data type
columns_fx = sp500_history_dataframe["Close"].str.split("$", expand=True)
# drop columns [0] asap
columns_fx.head(5)
```

```python
#naming column to Close
sp500_history_dataframe["Close"] = columns_fx[1]
sp500_history_dataframe.head()
```

```python
# ...changing data_type: from object to float
sp500_history_dataframe["Close"] = sp500_history_dataframe.Close.astype('float')                                
sp500_history_dataframe.dtypes
```

```python
# Calculate Daily Returns
sp500_returns_dataframe = sp500_history_dataframe.pct_change()
sp500_returns_dataframe.head()
```

```python
# Drop nulls
# Count nulls
sp500_returns_dataframe.isnull()
sp500_returns_dataframe.isnull().sum()
```

```python
# Drop nulls
sp500_returns_dataframe.dropna(inplace=True)
sp500_returns_dataframe.isnull().sum()
sp500_returns_dataframe.head(5)
```

```python
# Rename Column
sp500_returns_dataframe = sp500_returns_dataframe.rename(columns={
    "Close": "S&P 500"})
sp500_returns_dataframe.head()
```

```python
# Combine Whale, Algorithmic, and S&P 500 Returns
```

```python
# Concatenate whale_returns, algo_returns, and sp500_returns into a single dataframe 
whale_algo_sp500 = pd.concat([whale_returns_dataframe, algo_returns_dataframe, sp500_returns_dataframe], axis="columns", join="inner")
whale_algo_sp500.head(5)
```

```python
# Portfolio Analysis
# Performance
# Plot daily returns
whale_algo_sp500.plot(figsize=(10,6))
```

```python
# Plot cumulative returns
cumulative_returns = (1 + whale_algo_sp500).cumprod() - 1
cumulative_returns.plot(figsize = (10,5))
```

```python
### Reading the file the second stock
aapl_historical_dataframe = pd.read_csv(aapl_historical_csv, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
aapl_historical_dataframe.head(5)
```

```python
## Read the third stock Cost COST historical data
```

```python
cost_historical_csv = Path("cost_historical.csv")
```

```python
### Reading the file
cost_historical_dataframe = pd.read_csv(cost_historical_csv, index_col="Trade DATE", infer_datetime_format=True, parse_dates=True)
cost_historical_dataframe.head(5)
```

```python
# Concatenate all stocks GOOG, AAPL, and COST stocks in a DataFrame 
Concat_Stocks = pd.concat([goog_historical_dataframe, aapl_historical_dataframe, cost_historical_dataframe], axis="rows", join="inner")
Concat_Stocks.head(5)
```

```python
# Reset the index
#Concat_Stocks = Concat_Stocks.reset_index()
Concat_Stocks.head(5)
```

```python
# Pivot the Data so that the stock tickers are the columns, the dates are the index
# the values are the closing prices
Concat_Stocks = Concat_Stocks.pivot_table(index="Trade DATE", columns="Symbol")
Concat_Stocks.head(5)
```

```python
# Drop Nulls
Concat_Stocks.dropna(inplace=True)
Concat_Stocks.head(5)
```

```python
## Calculating daily returns and drop nulls
Concat_Stocks_returns = Concat_Stocks.pct_change()
Concat_Stocks_returns.dropna(inplace=True)
Concat_Stocks_returns.head(5)
```

```python
Concat_Stocks_returns.columns
```

```python
# Calculate the weighted returns for the portfolio assuming equal number of shares for each stock
```

```python
goog_weight = 1/3
aapl_weight = 1/3
cost_weight = 1/3
portfolio_returns = (goog_weight * Concat_Stocks_returns["NOCP"]["GOOG"]) + (aapl_weight * Concat_Stocks_returns["NOCP"]["AAPL"]) + (cost_weight * Concat_Stocks_returns["NOCP"]["COST"])
portfolio_returns.head(5)
```

```python
# Join all portfolio returns in one DataFrame
```

```python
all_portfolios_return = pd.concat([whale_algo_sp500, portfolio_returns], axis="columns", join="inner")
all_portfolios_return.dropna(inplace = True)
all_portfolios_return.tail(5)
```

```python
# Naming the customized portfolio: Custom

all_portfolios_return.rename(columns={0: "Custom"},inplace=True)
```
